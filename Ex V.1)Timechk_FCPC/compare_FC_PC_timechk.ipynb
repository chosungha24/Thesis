```python
import pickle
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams.update({'font.size': 20})

import os
os.environ["CUDA_DEVICE_ORDER"]="PCI_BUS_ID"
os.environ["CUDA_VISIBLE_DEVICES"]="1"
torch.set_default_dtype(torch.float64)
```

```python
class DNN(nn.Module):
    def __init__(self, neurons, activation):

        super(DNN, self).__init__()
        self.activation = activation
        self.n_layers = len(neurons)
        self.dense = {}
        self.nn_layers = nn.ModuleList()

        for i in range(self.n_layers-1):
            self.dense[str(i)] = nn.Linear(neurons[i], neurons[i+1])
            nn.init.xavier_uniform_(self.dense[str(i)].weight)
            nn.init.zeros_(self.dense[str(i)].bias)
            self.nn_layers.append(self.dense[str(i)])
            
    def forward(self, *args):
        x = torch.cat(args,1)
        activation = self.activation
        for i in range(self.n_layers-2):
            x = activation(self.dense[str(i)](x))
        i = self.n_layers-2
        return self.dense[str(i)](x)
```

```python
device='cuda'

dim_x = 1
dim_z = 9
net_size_G = [dim_x+dim_z,64,64,1]
net_size_Gx = [dim_x,64,64,10]
net_size_Gz = [dim_z,64,64,10]

G = DNN(net_size_G, activation=torch.tanh)
Gx = DNN(net_size_Gx, activation=torch.tanh)
Gz = DNN(net_size_Gz, activation=torch.tanh)

G.to(device); Gx.to(device); Gz.to(device)

def oPCN(X,Z):
    outx = Gx(X)
    outz = Gz(Z)
    return torch.matmul(outx,outz.T)

def oFCN(X,Z):
    xsize=X.shape[0]
    zsize=Z.shape[0]
    
    X = X.repeat([1,zsize]).reshape(-1,dim_x)
    Z = Z.repeat([xsize,1]).reshape(-1,dim_z)
    return G(X,Z)

PCN = torch.compile(oPCN)
FCN = torch.compile(oFCN)
```

```python
torch.manual_seed(0)

mean_time_FCN = {}
mean_time_PCN = {}
std_time_FCN = {}
std_time_PCN = {}

m = 500
N_time = [1000, 2000, 3000, 4000]
nloops=100000

start = torch.cuda.Event(enable_timing=True)
end = torch.cuda.Event(enable_timing=True)
for N in N_time:
    print(f"N={N}")
    x = torch.linspace(0,1,m, device=device).reshape(-1,1)
    z = torch.randn([N,dim_z], device=device)
    
    time_list=[]
    for _ in range(nloops):
        start.record()
        FCN(x,z)
        end.record()
        torch.cuda.synchronize()
        time_list.append(start.elapsed_time(end))
    mean_time_FCN[N]=np.array(time_list).mean()
    std_time_FCN[N]=np.array(time_list).std()
    
    time_list=[]
    for _ in range(nloops):
        start.record()
        PCN(x,z)
        end.record()
        torch.cuda.synchronize()
        time_list.append(start.elapsed_time(end))
    mean_time_PCN[N]=np.array(time_list).mean()
    std_time_PCN[N]=np.array(time_list).std()

    print(mean_time_FCN[N], mean_time_PCN[N])
```

```python
res_mfixed = {}
res_mfixed['FCNmean']=mean_time_FCN
res_mfixed['PCNmean']=mean_time_PCN
res_mfixed['FCNstd']=std_time_FCN
res_mfixed['PCNstd']=std_time_PCN

with open('results/time_m_fixed.pickle', 'wb') as f:
    pickle.dump(res_mfixed, f, pickle.HIGHEST_PROTOCOL)
```

```python
colors = [plt.cm.Paired(i) for i in range(20)]
cmap = {'FCN':colors[1], 'PCN':colors[5], 'ref':'k'}

fig = plt.figure(figsize=(8,8))
axes = [plt.subplot(1,1,1)]

axes[0].plot(range(1,len(N_time)+1), list(mean_time_FCN.values()), color=cmap['FCN'], label='FCN', marker='^', markersize=10)
axes[0].plot(range(1,len(N_time)+1), list(mean_time_PCN.values()), color=cmap['PCN'], label='PCN', marker='s', markersize=10)
axes[0].set_xticks([i for i in range(1,len(N_time)+1)])
axes[0].set_xticklabels(N_time)
axes[0].set_xlabel('N')
axes[0].legend()
```
