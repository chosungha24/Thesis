```python
import pickle
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams.update({'font.size': 20})
```

```python
with open('results/time_m_fixed.pickle', 'rb') as f:
    res_mfixed = pickle.load(f)
    
with open('results/time_N_fixed.pickle', 'rb') as f:
    res_Nfixed = pickle.load(f)
```

```python
colors = [plt.cm.Paired(i) for i in range(20)]
cmap = {'FCN':colors[1], 'PCN':colors[5], 'ref':'k'}

m_time = [100, 200, 300, 400]
N_time = [1000, 2000, 3000, 4000]


fig = plt.figure(figsize=(16,8))
axes = [plt.subplot(1,2,1), plt.subplot(1,2,2)]

mean_time_FCN = res_mfixed['FCNmean']
mean_time_PCN = res_mfixed['PCNmean']
axes[0].plot(range(1,len(N_time)+1), list(mean_time_FCN.values()), color=cmap['FCN'], label='FCN', marker='^', markersize=10)
axes[0].plot(range(1,len(N_time)+1), list(mean_time_PCN.values()), color=cmap['PCN'], label='PCN', marker='s', markersize=10)
axes[0].set_xticks([i for i in range(1,len(N_time)+1)])
axes[0].set_xticklabels(N_time)
axes[0].set_xlabel('N')
axes[0].set_ylabel('time (ms)')
axes[0].set_title('m=500')
axes[0].legend()

mean_time_FCN = res_Nfixed['FCNmean']
mean_time_PCN = res_Nfixed['PCNmean']
axes[1].plot(range(1,len(N_time)+1), list(mean_time_FCN.values()), color=cmap['FCN'], label='FCN', marker='^', markersize=10)
axes[1].plot(range(1,len(N_time)+1), list(mean_time_PCN.values()), color=cmap['PCN'], label='PCN', marker='s', markersize=10)
axes[1].set_xticks([i for i in range(1,len(N_time)+1)])
axes[1].set_xticklabels(m_time)
axes[1].set_xlabel('m')
axes[1].set_ylabel('time (ms)')
axes[1].set_title('N=5000')
axes[1].legend()

fig.savefig('images/FCPC_timechk.pdf', bbox_inches='tight')
```
